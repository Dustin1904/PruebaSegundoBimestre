```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
archivo=pd.read_csv("Electric_Vehicle_Population_Data.csv") #Leer el archivo de la base de datos
archivo=archivo.drop_duplicates() # Eliminar registros duplicados
archivo=archivo.dropna() # Eliminar registros vacios
archivo=archivo.fillna(0) # Reemplazar datos vacios por un dato especifico, en este caso cero
archivo # Imprimimos el data frame
```

```python
archivo['Model Year'].value_counts().plot(kind="bar")
```

```python
busquedas = ["Yakima", "Chelan", "Lewis","Pierce","King"]
county = archivo['County'].value_counts()
cantidad = county[busquedas]
cantidad.plot(kind="bar")
```

```python
busquedas1 = ["TESLA", "FORD", "CHEVROLET"]
marca = archivo['Make'].value_counts()
cantidad = marca[busquedas1]
cantidad.plot(kind="pie")
```

```python
estados=archivo['State'].value_counts().plot(kind="pie")
```

```python
busquedas = ["MODEL Y", "BOLT EV", "FUSION","BOLT EV"]
modelo = archivo['Model'].value_counts()
cantidad=modelo[busquedas]
cantidad.plot(kind="pie")
```
